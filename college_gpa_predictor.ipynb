```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('sat_gpa_v1.csv')
data
```

```python
data.info()
```

```python
scatter = sns.scatterplot(data=data, x='sat_sum', y='fy_gpa')
scatter.set_xlabel('SAT Summary')
scatter.set_ylabel('Fiscal Year GPA')
```

```python
lmplot = sns.lmplot(data=data, x='sat_sum', y='fy_gpa')
lmplot.set(xlabel='SAT Summary', ylabel='Fiscal Year GPA')
```

### Scatter Plot for the analysis, here we see a weak correlation between SAT scores and Fiscal Year GPA. This supports a theory that there are people who just take tests better than others.

```python
## regression model
import statsmodels.api as sm
```

```python
model = sm.OLS(data['sat_sum'], sm.add_constant(data['fy_gpa'])).fit()
print(model.summary())
```

### OLS Model for the linear regression

#### according to this regression model...

```python


lmplot = sns.lmplot(data=data, x='sat_sum', y='fy_gpa', scatter_kws={'alpha': 0.3, 'color': 'orange'},
                    line_kws={'alpha': 0.7, 'color':'lightblue'})
lmplot.set(xlabel='SAT Summary', ylabel='Fiscal Year GPA')
```
